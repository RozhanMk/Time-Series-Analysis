```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AirPassengers.csv', index_col="Month", parse_dates=True)
display(df.head())
df.plot()
```

```python
from pmdarima import auto_arima

NS_test = 12
train_data = df["#Passengers"][:-NS_test]
test_data = df["#Passengers"][-NS_test:]

train_idx = df.index <= train_data.index[-1]
test_idx = df.index > train_data.index[-1]
```

```python
model = auto_arima(train_data, trace=True, start_p=1, start_q=1, seasonal=True,m=12, suppress_warnings=True)
```

```python
model.summary()
```

```python
model.plot_diagnostics(figsize=(10,8))
plt.show()
```

```python
fc, confint = model.predict(n_periods=NS_test, return_conf_int=True)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['#Passengers'], label='real values')
plt.plot(df.index[test_idx], fc, label='forecast')
plt.fill_between(df.index[test_idx], confint[:, 0], confint[:, 1], alpha=0.1)
plt.title('Real vs Forecast Data')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
rmse_arima = np.sqrt(mean_squared_error(test_data, fc))
print(f'RMSE (Auto_ARIMA): {rmse_arima}')

naive_prediction = np.full(NS_test, train_data.iloc[-1])

# Calculate RMSE for the Naive model
rmse_naive = np.sqrt(mean_squared_error(test_data, naive_prediction))
print(f'RMSE (Naive Model): {rmse_naive}')

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['#Passengers'], label='real values')
plt.plot(df.index[test_idx], fc, label='Auto_ARIMA forecast')
plt.plot(df.index[test_idx], naive_prediction, label='Naive forecast')
plt.fill_between(df.index[test_idx], confint[:, 0], confint[:, 1], alpha=0.1)
plt.title('Real vs Forecast Data (Auto_ARIMA vs Naive)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('all_stocks_5yr.csv', index_col="date", parse_dates=True)
display(df.head())
```

```python
df = df.loc[df['Name'] == 'AAPL']
df[["close"]].plot()
```

```python
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error

# Split data into training and testing sets
NS_test = 30
train_data = df["close"][:-NS_test]
test_data = df["close"][-NS_test:]

train_idx = df.index <= train_data.index[-1]
test_idx = df.index > train_data.index[-1]

model = auto_arima(train_data, trace=True, seasonal=False, suppress_warnings=True, maxiter=10)
model.summary()
```

```python
fc, confint = model.predict(n_periods=NS_test, return_conf_int=True)

rmse_arima = np.sqrt(mean_squared_error(test_data, fc))
print(f'RMSE (Auto_ARIMA): {rmse_arima}')

naive_prediction = np.full(NS_test, train_data.iloc[-1])

rmse_naive = np.sqrt(mean_squared_error(test_data, naive_prediction))
print(f'RMSE (Naive Model): {rmse_naive}')
```

As we see, naive forecast works better than our auto arima.

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index[test_idx], df.iloc[test_idx]['close'], label='real values')
plt.plot(df.index[test_idx], fc, label='Auto_ARIMA forecast')
plt.plot(df.index[test_idx], naive_prediction, label='Naive forecast')
plt.fill_between(df.index[test_idx], confint[:, 0], confint[:, 1], alpha=0.1)
plt.title('Real vs Forecast Data (Auto_ARIMA vs Naive)')
plt.xlabel('Date')
plt.ylabel('Close value')
plt.legend()
plt.show()
```

## Let's find optimal ARIMA using ACF and PACF

```python
def log_return(df):
  df["log_return"] = np.log(df["close"]) - np.log(df["close"].shift(1))
  return df
```

```python
df = log_return(df).dropna()
df[["log_return"]].plot()
```

We get Log return by differnciating between log prices. This is a solution to get a stationary time series. You can test the stationary trait by dicky fuller test.
Using the result below, we can see that p_value < 0.05. It means log return is stationary.

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
  result = adfuller(series.dropna())
  print('ADF Statistic: %f' % result[0])
  print('p-value: %f' % result[1])
  print('Critical Values:')
  for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

adf_test(df['log_return'])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df["log_return"])
plot_pacf(df["log_return"])
plt.show()
```

p and q in ARMA model based on PACF and ACF will be 0. ARIMA model is (0, 1, 0). It means this stock follows a random walk and cannot be predicted.
